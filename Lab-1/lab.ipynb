# Lab 01: Convolution Operation

```python
import cv2

from matplotlib import pyplot as plt
```

## Reading an RGB image with OpenCV

```python
PATH = "images/lena.png"
IMG = cv2.imread(PATH, cv2.IMREAD_COLOR)

plt.imshow(cv2.cvtColor(IMG, cv2.COLOR_BGR2RGB))
plt.show()
```

## Maximum and Minimum Intensity and Image Dimensions

```python
print("Maximum Intensity: ", IMG.max())
print("Minimum Intensity: ", IMG.min())
print("Mean Intensity: ", IMG.mean())
print("Standard Deviation: ", IMG.std())

print("Shape: ", IMG.shape)
print("Size: ", IMG.size)
print("Type: ", IMG.dtype)
```

## Convert RBG Image to Gray Scale Image

```python
gray_img = cv2.cvtColor(IMG, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_img, cmap="gray")
plt.show()
```

## Reading Image as Grayscale


```python
gray_img = cv2.imread(PATH, cv2.IMREAD_GRAYSCALE)

plt.imshow(gray_img, cmap="gray")
plt.show()
```

## Operations on Grayscale Images

Images can be thought of as matrices of pixels. Each pixel is a number between 0 and 255.
So we can perform operations on the pixels.

```python
row, col = IMG.shape[:2]

for i in range(row):
    for j in range(col):
        if gray_img[i, j] % 2 == 0:
            gray_img[i, j] = 255

plt.imshow(gray_img, cmap="gray")
plt.show()
```

```python
gray_img = cv2.imread(PATH, cv2.IMREAD_GRAYSCALE)
row, col = IMG.shape[:2]


for i in range(row):
    for j in range(col):
        if gray_img[i, j] % 2 == 0:
            gray_img.itemset((i, j), 0)

plt.imshow(gray_img, cmap="gray")
plt.show()
```
