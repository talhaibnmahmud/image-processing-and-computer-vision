# Histogram

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img: cv2.Mat = cv2.imread('images/lena.tif', 0)

plt.imshow(img, cmap='gray')
plt.show()

print('Image shape: ', img.shape)
print('Data type: ', img.dtype)
print('Min value: ', img.min())
print('Max value: ', img.max())
```

```python
row, col = img.shape[:2]
intensity = np.zeros(256, dtype=np.uint8)

print(intensity)
```

```python
for i in range(row):
    for j in range(col):
        intensity[img[i, j]] += 1

print(intensity)
print(intensity.sum())
```

```python
plt.hist(img.ravel(), 256, [0, 256])
plt.title('Histogram of the input image')
plt.show()
```

```python
plt.plot(intensity)
plt.title('Calculated histogram')
plt.show()
```

```python
histogram = np.divide(intensity, row * col)

plt.plot(histogram)
plt.title('Normalized histogram')
plt.show()
```

```python
hist = np.zeros(256)
s = np.zeros(256)

sum_hist = 0
for i in range(256):
    sum_hist += histogram[i]
    hist[i] = sum_hist + histogram[i]
    s[i] = round(hist[i] * 255)

print(hist)
print(s)
```

```python
plt.plot(hist)
plt.title('Histogram of the input image')
plt.show()
```

```python
from copy import deepcopy as dpc

img_inp=cv2.imread('images/lena.tif', cv2.IMREAD_GRAYSCALE)
a_min = np.min(img_inp)
a_max = np.max(img_inp)

for i in range (img_inp.shape[0]):
    for j in range(img_inp.shape[1]):
        img_inp[i][j] = (((img_inp[i][j]-a_min)/(a_max-a_min))*255)
img_inp = img_inp.astype(np.uint8)
img = dpc(img_inp)

print(a_max)
print(a_min)
```

```python
each_intensity = [0 for _ in range (256)]
each_intensity_out = [0 for _ in range (256)]
histogram = [0.0 for _ in range (256)]
histogram_out = [0.0 for _ in range (256)]
Hist = [0.0 for _ in range(256)]
Hist_out = [0.0 for _ in range(256)]
s = [0.0 for _ in range (256)]
```

```python
image_size=img.shape[0]*img.shape[1]
print(image_size)
```

```python
##histogram calculation for input
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        x = img.item(i,j)
        each_intensity[x] += 1

print(each_intensity)
plt.plot(each_intensity)
plt.show()
```

```python
for i in range(256):
    histogram[i] = each_intensity[i] / image_size

plt.plot(histogram)
plt.show()
```

```python
## CDF calculation for input
sum_hist = 0

for i in range(256):
    sum_hist = sum_hist + histogram[i]
    Hist[i] = sum_hist + histogram[i]
    s[i]=round(sum_hist*255)

print(s)

plt.plot(Hist)
plt.show()
```

```python
plt.plot(s)
plt.show()
```

```python

## output update
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        x=img[i][j]
        img[i][j]=s[x]

##CDF of output
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        x = img.item(i,j)
        each_intensity_out[x] += 1

print(each_intensity_out)
plt.plot(each_intensity_out)
plt.show()
```

```python
for i in range(256):
    histogram_out[i] = each_intensity_out[i]

sum_hist = 0

for i in range(256):
    sum_hist = sum_hist + histogram_out[i]
    Hist_out[i] = sum_hist

cv2.imshow("input image",img_inp)
plt.hist(img_inp.ravel(),256,[0,256])
plt.title("Input image")
plt.show()

plt.plot(Hist)
plt.title("Input historam")
plt.show()

cv2.imshow("Output image",img)
plt.hist(img.ravel(),256,[0,256])
plt.title("Output image original")
plt.show()

plt.plot(Hist_out)
plt.title("Output histogram")
plt.show()
```
